```python
df1=df.loc[df["Species_name"]=="Setosa"]
df2=df.loc[df["Species_name"]=="Versicolor"]
df3=df.loc[df["Species_name"]=="Verginica"]
```

```python
valeurs_moyennes1=[df1.Petal_width.mean(),df1.Petal_length.mean(),df1.Sepal_width.mean(),df1.Sepal_length.mean()]
valeurs_moyennes2=[df2.Petal_width.mean(),df2.Petal_length.mean(),df2.Sepal_width.mean(),df2.Sepal_length.mean()]
valeurs_moyennes3=[df3.Petal_width.mean(),df3.Petal_length.mean(),df3.Sepal_width.mean(),df3.Sepal_length.mean()]
ecarts_types1=[df1.Petal_width.values.std(),df1.Petal_length.values.std(),df1.Sepal_width.values.std(),df1.Sepal_length.values.std()]
ecarts_types2=[df2.Petal_width.values.std(),df2.Petal_length.values.std(),df2.Sepal_width.values.std(),df2.Sepal_length.values.std()]
ecarts_types3=[df3.Petal_width.values.std(),df3.Petal_length.values.std(),df3.Sepal_width.values.std(),df3.Sepal_length.values.std()]
```

```python

ecarts_types_norm1 = [1.96*ec for ec in ecarts_types1]
ecarts_types_norm1
bar_labels = ['Petal_width','Petal_length','Sepal_width','Sepal_length']
fig, ax = plt.subplots()
x_position = list(range(len(bar_labels)))
plt.bar(x_position, valeurs_moyennes1, yerr = ecarts_types_norm1)
```
