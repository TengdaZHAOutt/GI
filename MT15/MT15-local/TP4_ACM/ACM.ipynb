```python
# Importation des librairies
import pandas as pd
import numpy as np
from fanalysis.mca import MCA
import matplotlib.pyplot as plt
```

```python
# On convertit les données en format matriciel (numpy array)
X = df.values
```

## Application de la méthode ACM

```python
# On crée une instance de la classe MCA, en lui passant ici des labels pour les lignes et les colonnes
# Par défaut, on effectue une ACM en retenant toutes les composantes
# Pour retenir un nombre prédéfini (ex. n=2), il suffit de rajouter l'argument n_components=2
ACM = MCA(row_labels=df.index.values, var_labels=df.columns.values)
```

```python
# On estime le modèle en appliquant la méthode fit de la classe MCA sur le jeu de données.
ACM.fit(X)
```

L'exécution de la méthode ACM.fit(X) a entrainé le calcul de plusieurs statistiques dont :

* ACM.eig_ : les valeurs propres, en pourcentage, et en poucentage cumulé
* ACM.row_coord_ : les coordonnées des points lignes
* ACM.col_coord_ : les coordonnées des points colonnes

## Analyse des valeurs propres

L'attribut ACM.eig_ comporte :

* en 1ère ligne : les valeurs propres en valeur absolue
* en 2ème ligne : les valeurs propres en pourcentage de la variance totale
* en 3ème ligne : les valeurs propres en pourcentage cumulé de la variance totale

```python
# Affichage des valeurs propres
valeurs_propres=ACM.eig_[0]
print("Les valeurs propres sont : ", valeurs_propres)
```

```python
# Pourcentage de l'inertie exprimée par chaque axe
print("Le poucentage d'inertie expérimée par chaque axe : ", ACM.eig_[1])
```

```python
# Pourcentage cumulé de l'inertie totale
print("Les poucentages d'inertie cumulée : ", ACM.eig_[2])
```

```python
# Théoriquement
print("Le nombre maximal de dimensions est : K - p = ", 3*5+2*2-7)
print("L'inertie total est : K/p - 1 = ", (3*5+2*2)/7-1)
```

```python
# Vérification avec les résultats obtenus
print("Le nombre maximal de dimensions obtenu = ", valeurs_propres.shape)
print("L'inertie totale obtenue = ",valeurs_propres.sum())
```

## Selection des facteurs pertinents 

Un seuil de sélection possible : on sélectionne les facteurs dont les v.p. sont supérieures à la moyenne des v.p.

```python
print("La moyenne des valeurs propres : ", np.average(ACM.eig_[0]))
print("La moyenne des valeurs propres est aussi égale à 1/p : ", 1/7)
print("On peut retenir les dimensions qui possèdent une valeur propre supérieure à la moyenne des valeurs propres. On retient donc", sum(ACM.eig_[0]>1/7), "dimensions.")
```

Utilisation d'un indicateur corrigé : **formule de taux d’inertie corrig (Benzecri 1979)**


$$\lambda_{h}^{'} = \left(\frac{p}{p-1} \times (\lambda_{h}-\frac{1}{p})\right)^{2}$$

p est le nombre de variables.

```python
# Formule de taux d’inertie corrigé (Benzecri 1979)
vp_corrigés = ((ACM.eig_[0]-1/7)*7/6)**2
```

```python
# Pourcentages d'inertie expliquée
100*(vp_corrigés[:4]/vp_corrigés[:4].sum())
```

On refait le calcul en retenant uniquement 4 dimensions

```python
# Instanciation
ACM4 = MCA(row_labels=df.index.values, var_labels=df.columns.values,n_components=4)
# Calcul
ACM4.fit(X)
```

```python
# Représentation des valeurs propres
ACM4.plot_eigenvalues()
```

```python
# Représentation des valeurs propres en poucentage d'inertie
ACM4.plot_eigenvalues(type="percentage")
```

```python
# Représentation des valeurs propres en pourcentage d'inertie cumulé
ACM4.plot_eigenvalues(type="cumulative")
```

Lors de l'exécution de la méthode ACM4.fit(X), les statistiques suivantes ont été calculées :

* ACM4.row_contrib_ : contributions des points lignes à la variance de l'axe
* ACM4.col_contrib_ : contributions des points colonnes à la variance de l'axe
* ACM4.row_cos2_ : cosinus carrés des points lignes
* ACM4.col_cos2_ : cosinus carrés des points colonnes
